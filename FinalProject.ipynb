```python
from sklearn.impute import SimpleImputer

num_imp = SimpleImputer(strategy="mean")
df["age"] = num_imp.fit_transform(df[["age"]]).ravel()

cat_imp = SimpleImputer(strategy="most_frequent")
df["employment_type"] = cat_imp.fit_transform(df[["employment_type"]]).ravel()
```

```python
df["income_missing"] = df["annual_income"].isnull().astype(int)

null_mask = df["annual_income"].isnull()
num_nulls = null_mask.sum()

sampled_values = df["annual_income"].dropna().sample(num_nulls, replace=True)

df.loc[null_mask, "annual_income"] = sampled_values.values
```

```python
from sklearn.impute import KNNImputer

knn = KNNImputer(n_neighbors=5)
df[["annual_income","loan_amount","credit_score"]] = knn.fit_transform(
    df[["annual_income","loan_amount","credit_score"]]
)
```

```python
Q1 = df["loan_amount"].quantile(0.25)
Q3 = df["loan_amount"].quantile(0.75)
IQR = Q3 - Q1

df = df[(df["loan_amount"] > Q1 - 1.5*IQR) &
        (df["loan_amount"] < Q3 + 1.5*IQR)]
```

```python
from scipy.stats.mstats import winsorize

df["credit_score"] = winsorize(df["credit_score"], limits=[0.05,0.05])
```

```python
df["year"] = pd.to_datetime(df["join_date"]).dt.year
df["month"] = pd.to_datetime(df["join_date"]).dt.month
df["weekday"] = pd.to_datetime(df["join_date"]).dt.weekday
```

```python
edu_map = {"Primary":1,"Secondary":2,"Graduate":3,"Post-Graduate":4}
df["education_level"] = df["education_level"].map(edu_map)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["gender"] = le.fit_transform(df["gender"])
```

```python
df = pd.get_dummies(df, columns=["region","loan_purpose"], drop_first=True)
```

```python
df["income_bin"] = pd.qcut(df["annual_income"], 4, labels=False)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[["annual_income","loan_amount"]] = scaler.fit_transform(
    df[["annual_income","loan_amount"]]
)
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
df[["spending_ratio"]] = mm.fit_transform(df[["spending_ratio"]])
```

```python
df["debt_to_income"] = df["loan_amount"] / df["annual_income"]
df["avg_transactions"] = df["transaction_count"] / 6
```

```python
import numpy as np

df["log_income"] = np.log1p(df["annual_income"])
df["sqrt_spending"] = np.sqrt(df["spending_ratio"])
```
